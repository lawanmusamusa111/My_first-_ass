```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
Data = pd.read_csv("/content/petrol_consumption.csv")
Data
```

EXPLOTARY DATA ANALYSIS (EDA
)

```python
#HISTOGRAM
Data.hist(figsize = (3,7))
plt.tight_layout()
plt.show()
```

```python
#bxplot chart with seaborn
sns.boxplot( x = "Paved_Highways", y ="Petrol_Consumption", data = Data )
plt.title("The petrol distribution of the data set")
plt.show()
```

```python
#scatter plot
sns.scatterplot(x = "Paved_Highways", y ="Petrol_Consumption", hue ='Average_income', data = Data)
plt.title("A scatterplot for the two diff column")
plt.show()
```

```python
# the heatmap
sns.heatmap(Data.corr(), annot = True, cmap ="coolwarm")
plt.title("THE CORRELATION OF THE PETROLEUM COMSUMPTION")
plt.show()
```

```python
#barchart
sns.countplot( x ="Average_income", data = Data)
plt("BARCHART")
plt.show()
```

```python
#pie chart
plt.pie(Data["Average_income"].value_counts(), labels= Data["Average_income"].unique())
plt.title("PIes showing fuel income")
plt.show()
```
